```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 10, 100)
y = np.sin(x)
```

```python
plt.plot(x, y, color="red", linestyle=":")
plt.xlabel("Numbers (1 - 10)")
plt.ylabel("Sin of X")
plt.title("Numbers vs Sin")
plt.show()
```

```python
fig, (ax1, ax2) = plt.subplots(ncols=2, nrows=1, figsize=(12, 4), sharey=True)
ax1.plot(x, y)
ax1.set_xlabel("Numbers (1 - 10)")
ax1.set_ylabel("Sin of X")
ax1.set_title("Numbers VS Sin")

ax2.plot(x, y, color="red")
ax2.set_xlabel("Numbers (1 - 10)")
ax2.set_ylabel("Sin of X")
ax2.set_title("Numbers VS Sin")
plt.show()
```

```python
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
```

```python
plt.plot(x, y1, label="SIN")
plt.plot(x, y2, label="COS")
plt.xlabel("Numbers (1 - 10)")
plt.ylabel("SIN/COS of X")
plt.title("NUMBERS VS SIN/COS")
plt.legend(loc=(0.6, 0.1))
plt.show()
```

```python
categories = ['A', 'B', 'C', 'D']
values = [10, 15, 12, 8]
```

```python
plt.bar(categories, values, color="purple")
plt.xlabel("Categories")
plt.ylabel("Values")
plt.show()
```
