```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 10, 100)
y = np.sin(x)
```

```python
plt.plot(x, y, color="red", linestyle=":")
plt.xlabel("Numbers (1 - 10)")
plt.ylabel("Sin of X")
plt.title("Numbers vs Sin")
plt.show()
```

```python
fig, (ax1, ax2) = plt.subplots(ncols=2, nrows=1, figsize=(12, 4), sharey=True)
ax1.plot(x, y)
ax1.set_xlabel("Numbers (1 - 10)")
ax1.set_ylabel("Sin of X")
ax1.set_title("Numbers VS Sin")

ax2.plot(x, y, color="red")
ax2.set_xlabel("Numbers (1 - 10)")
ax2.set_ylabel("Sin of X")
ax2.set_title("Numbers VS Sin")
plt.show()
```

```python
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
```

```python
plt.plot(x, y1, label="SIN")
plt.plot(x, y2, label="COS")
plt.xlabel("Numbers (1 - 10)")
plt.ylabel("SIN/COS of X")
plt.title("NUMBERS VS SIN/COS")
plt.legend(loc=(0.6, 0.1))
plt.show()
```

```python
categories = ['A', 'B', 'C', 'D']
values = [10, 15, 12, 8]
```

```python
plt.bar(categories, values, color="purple")
plt.xlabel("Categories")
plt.ylabel("Values")
plt.show()
```

```python
data = np.random.normal(size=1000)
plt.hist(data, bins=30)
plt.show()
```

```python
x = np.random.rand(50)
y = np.random.rand(50)

plt.scatter(x, y)
plt.show()
```

```python
labels = ['A', 'B', 'C', 'D']
sizes = [25, 35, 20, 20]

plt.pie(sizes, labels=labels, autopct='%.1f%%', radius=1.1, counterclock=False, colors=["red", "green", "purple", "cyan"],
        explode=[0 , 0.2 , 0 , 0])

plt.show()
```

```python
data = np.random.normal(size=1000)

plt.boxplot(data, vert=False)
plt.show()
```

```python
plt.style.use("ggplot")
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
plt.figure(figsize=(14,7))
plt.plot(x, y1, color='red', linestyle='-', marker='o', label='sin(x)')
plt.plot(x, y2, color='blue', linestyle='--', marker='s', label='cos(x)')

plt.legend()
plt.show()
```

```python
%matplotlib inline

x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.figure(figsize=(14, 7))
plt.plot(x, y)
plt.xticks(color="red", rotation=45)
plt.yticks(color="green")
plt.xlabel("Numbers", color="red")
plt.ylabel("Sin of X", color="green")
plt.title("Number VS Sin", color="magenta")

plt.text(x=4.3, y=-0.75, s="Bottom", color="blue", fontsize="xx-large", fontweight="bold")
plt.annotate(text="Start", xy=(0, 0), xytext=(0.3, -0.1), arrowprops=dict(facecolor='green', shrink=0.05), 
            color="green", fontweight="semibold", fontsize="x-large"
            )
plt.savefig("img.png", dpi=300)
plt.show()
```
