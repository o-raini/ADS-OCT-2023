```python
data = {
    'A': [1, 2, 3],
    'B': [4, 5, 6],
    'C': [7, 8, 9]
}
df = pd.DataFrame(data)
df
```

```python
def calc(x, y):
    return x**y
```

```python
df["cube"] = df["C"].apply(calc, y=2)
```

```python
df
```

```python
df["B_cube"] = df["B"].map(calc, y=2)
df
```
